## Data Preparation

```python
import pandas as pd
import numpy as np
```

```python
df_can = pd.read_excel('datasets/Canada.xlsx',sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

```python
df_can.tail()
```

```python
df_can.info()
```

```python
df_can.shape
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
```

```python
df_can.rename(columns={'OdName': 'Country', 'AreaName': 'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.head()
```

```python
df_can.describe()
```

```python
df_can[['Country', 2000, 2001, 2002, 2003]]
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head(3)
```

```python
df_can.loc['Indonesia', 2013]
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980, 2014)))
print(years)
```

## Data Visualization

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```

```python
df_can.head()
```

## Area Plot

```python
df_top5.plot(kind='area',
            stacked=False,
            figsize=(20, 10),
            alpha=0.25)
plt.title('Imigration Tren Top 5 Country')
plt.xlabel('Years')
plt.ylabel('Number of Imigrants')

plt.show()
```

## Bar Charts

```python
df_iceland = df_can.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar',
         figsize=(20,10))

plt.title('Iceland Immigrants to Canada from 1980 - 2013')
plt.xlabel('Number of Years')
plt.ylabel('Number of Imigrants')

plt.show()
```

```python
df_iceland.plot(kind='bar',
         figsize=(10,6))

plt.title('Iceland Immigrants to Canada from 1980 - 2013')
plt.xlabel('Number of Years')
plt.ylabel('Number of Imigrants')
plt.annotate('',                      # s: str. will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )
# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis', # text to display
             xy=(28, 30),                    # start the text at at point (year 2008 , pop 30)
             rotation=72.5,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )
plt.show()
```

## Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie')
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(20,10),
                           autopct='%1.1f%%',
                            startangle = 90,
                            shadow=True
                           )
plt.title('Inigration to Canada by Continent (1980 - 2013)')
plt.axis('equal')

plt.show()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0.1, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                           figsize=(15,6),
                           autopct='%1.1f%%',
                            startangle = 90,
                            shadow=True,
                            colors=color_list,
                            explode=explode_list,
                            labels=None,
                            pctdistance=1.15
                           )
plt.title('Inigration to Canada by Continent (1980 - 2013)', y=1.15)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0.1, 0.1, 0.1]

df_continents['2013'].plot(kind='pie',
                           figsize=(15,6),
                           autopct='%1.1f%%',
                            startangle = 90,
                            shadow=True,
                            colors=color_list,
                            explode=explode_list,
                            labels=None,
                            pctdistance=1.15
                           )
plt.title('Inigration to Canada by Continent (2013)', y=1.15)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## Subplots

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: Box plot
df_CI.plot(kind='box', color='salmon', vert=False, figsize=(20, 6), ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India (1980 - 2013)')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

# Subplot 2: Line plot
df_CI.plot(kind='line', figsize=(20, 6), ax=ax1) # add to subplot 2
ax1.set_title ('Line Plots of Immigrants from China and India (1980 - 2013)')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')

plt.show()
```

## Scatter Plots

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace = True)
df_tot.columns = ['year', 'total']
```

```python
df_tot.head()
```

```python
df_tot.plot(kind='scatter', 
            x='year', 
            y='total', 
            figsize=(10, 6), 
            color='palevioletred')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
